## Concatenation and Addition of index in list

```python
list1=[1,2,3,4]
list2=[5,6,7,8]

print("Concatenation result: ",list1+list2)

sum_l=[]
for num1,num2, in zip(list1,list2):
  sum_l.append(num1+num2)
print("Addition of index result: ",sum_l)
```
