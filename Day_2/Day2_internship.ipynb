## **OOPs familiarisation**

```python
class student:
  def __init__(self,n,r,m,m2):
    self.name=n
    self.roll_no=r
    self.mark1=m
    self.mark2=m2
  def result(self):
    if(self.mark1>=40):print('pass')
    else: print('failed')
  def add_mrk(self):
    print('Sum: ',self.mark1 + self.mark2)

s1=student('abhishek',int(input("\nEnter the roll no: ")),int(input("\nEnter mark 1: ")),int(input("\nEnter mark 2: ")))
print(s1.mark1)
print(s1.name)
print(s1.roll_no)
student.result(s1)
student.add_mrk(s1)
```

**Approach 1** pass or fail

```python
s1=student('abc',4,80,35)
s2=student('def',5,30,45)
s3=student('ghi',6,50,25)
s4=student('jkl',7,90,43)

student_name=[s1,s2,s3,s4]
fail=[]
pass1=[]

for i in range(4):
  if student_name[i].mark1<40:
    fail.append(student_name[i].mark1)
  else:
    pass1.append(student_name[i].mark1)
print(fail)
print(pass1)
```

**Approach 2** for pass and fail list

```python
fail1=[]
pass12=[]

for x in student_name:
  if x.mark1>=40:
    pass12.append(x.name)
  else:
    fail1.append(x.name)
print(pass12)
print(fail1)
```

### By using class find out the ***area*** and ***perimeter*** of a circle

```python
class circle:
  def __init__(self,r):
    self.radius= r
    self.pie=3.14
  
  def area(self):
    print("Area of a circle: ",self.pie*self.radius*self.radius)         # area of circle = pi * r^2
  
  def perimeter(self):
    print("Perimeter of circle: ",2*self.pie*self.radius)                # perimeter of circle = 2 * pi * r

a=circle(int(input("Enter the radius of circle: ")))
circle.area(a)
circle.perimeter(a)
```
