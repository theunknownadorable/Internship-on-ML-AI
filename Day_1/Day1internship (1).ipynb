```python
a=int(input("Enter the number to be checked:"))
if a==1:
  print("Neither prime nor composite")
elif a>1:
  for i in range(2,a):
    if a % i == 0:
      flag=1
    else:
      flag=0
  if flag==1:
   print("The number is composite")
  else:
   print("The number is prime")
```
