### **Question 2**

```python
def fibro(n):
  if n <= 1:
    return n
  else:
    return (fibro(n - 1) + fibro(n - 2))

for i in range(10):
  print(fibro(i),end=' ')
```

### **Question 3**

```python
a=0
b=100
count=0    # count variable is used to find the number of prime nos
print("Prime numbers between 0 and 100 : ")
for num in range(a, b + 1):
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
                break
       else:
           print(num,end=" ")
           count=count+1
print(" \nTotal number of prime numbers between 0 and 100 are :" , count)
```

### **Question 7**

```python
# Function to get sum of digits 
def getSum(n):
  sum=0
  while(n>0):
    d=n%10
    sum=sum+d
    n=n//10
  return sum    

   
n = int(input("Enter you number: "))
print(getSum(n))
```
